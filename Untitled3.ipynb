```python
# colorgan.py  ─  Conditional GAN for image colourisation
# Inspired by https://towardsdatascience.com/colorizing-black-white-images-with-u-net-and-conditional-gan-a-tutorial-81b2df111cd8/

import os, random, pathlib
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from skimage.color import rgb2lab, lab2rgb

import torch, torch.nn as nn, torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torch.cuda.amp import GradScaler, autocast
import piq                                    # LPIPS / metrics (optional)

# ─────────── USER-TUNABLE FLAGS ───────────
DATA_DIR    = pathlib.Path('./places365SubsetRGB')
EPOCHS      = 5
INIT_BATCH  = 256
BATCH_AUTO  = True          # auto-shrink batch on OOM
CKPT_G      = pathlib.Path('gen.pth')
CKPT_D      = pathlib.Path('disc.pth')
λ_L1        = 100           # as in the pix2pix paper
LR_G        = 2e-4
LR_D        = 2e-4
DO_METRICS  = False         # CPU metrics after training
SHOW_IMAGES = True
# ──────────────────────────────────────────

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

device = (torch.device('cuda') if torch.cuda.is_available()
          else torch.device('mps') if torch.backends.mps.is_available()
          else torch.device('cpu'))
USE_AMP = device.type == 'cuda'
print(f'Using {device}   •   AMP={USE_AMP}')

# ---------- tiny dot-progress ----------
def progress(it, every=40, tag=''):
    for i, x in enumerate(it, 1):
        if i % every == 0: print(tag+'.', end='', flush=True)
        yield x
    if i >= every: print()

# ---------- safe autocast ----------
import contextlib
@contextlib.contextmanager
def amp():
    if USE_AMP:                       # real mixed precision on CUDA
        with autocast(dtype=torch.float16):
            yield
    else:                             # no-op on MPS / CPU
        yield

# ---------- Lab↔RGB (for metrics / display) ----------
def lab_to_rgb_tensor(lab_norm: torch.Tensor) -> torch.Tensor:
    L = lab_norm[:, :1] * 100.0
    ab = lab_norm[:, 1:] * 128.0
    lab = torch.cat([L, ab], 1)
    lab_np = lab.detach().cpu().numpy().transpose(0,2,3,1)
    rgb = [torch.tensor(lab2rgb(img), dtype=torch.float32).permute(2,0,1)
           for img in lab_np]
    return torch.stack(rgb).clamp(0,1)

# ---------- Dataset ----------
class ColorDS(Dataset):
    def __init__(self, root: pathlib.Path, size=256):
        self.files = [p for p in root.iterdir() if p.suffix.lower() in {'.jpg','.jpeg','.png'}]
        self.sz=size
    def __len__(self): return len(self.files)
    def __getitem__(self, idx):
        img = Image.open(self.files[idx]).convert('RGB')
        if img.size!=(self.sz,self.sz):
            img = img.resize((self.sz,self.sz), Image.LANCZOS)
        lab = rgb2lab(np.asarray(img,np.float32)/255.).astype(np.float32)
        L  = lab[..., [0]]/100.
        ab = lab[..., 1:]/128.
        return (torch.from_numpy(L).permute(2,0,1),
                torch.from_numpy(ab).permute(2,0,1))

def make_loader(split, bs):          # assumes DATA_DIR / train / val
    return DataLoader(ColorDS(DATA_DIR/split,256),
                      batch_size=bs, shuffle=(split=='train'),
                      pin_memory=(device.type=='cuda'), num_workers=0)

# ---------- Generator (U-Net) ----------
class ConvBlock(nn.Module):
    def __init__(self, in_c, out_c, norm=True, act=True):
        super().__init__()
        layers=[nn.Conv2d(in_c,out_c,4,2,1,bias=False)]
        if norm: layers.append(nn.BatchNorm2d(out_c))
        if act:  layers.append(nn.LeakyReLU(0.2,True))
        self.m=nn.Sequential(*layers)
    def forward(self,x): return self.m(x)

class DeconvBlock(nn.Module):
    def __init__(self, in_c, out_c, dropout=False):
        super().__init__()
        layers=[nn.ConvTranspose2d(in_c,out_c,4,2,1,bias=False),
                nn.BatchNorm2d(out_c),
                nn.ReLU(True)]
        if dropout: layers.append(nn.Dropout(0.5))
        self.m=nn.Sequential(*layers)
    def forward(self,x): return self.m(x)

class UNetG(nn.Module):
    def __init__(s):
        super().__init__()
        # encoder
        s.e1=ConvBlock(1,64, norm=False)        # (bs,64,128,128)
        s.e2=ConvBlock(64,128)                  # (64,64,64)
        s.e3=ConvBlock(128,256)                 # (32,32)
        s.e4=ConvBlock(256,512)                 # (16,16)
        s.e5=ConvBlock(512,512)                 # (8,8)
        s.e6=ConvBlock(512,512)                 # (4,4)
        s.e7=ConvBlock(512,512)                 # (2,2)
        s.e8=ConvBlock(512,512,norm=False, act=False)   # (1,1)
        # decoder
        s.d1=DeconvBlock(512,512,dropout=True)  # (2,2)
        s.d2=DeconvBlock(1024,512,dropout=True) # (4,4)
        s.d3=DeconvBlock(1024,512,dropout=True) # (8,8)
        s.d4=DeconvBlock(1024,512)              # (16,16)
        s.d5=DeconvBlock(1024,256)              # (32,32)
        s.d6=DeconvBlock(512,128)               # (64,64)
        s.d7=DeconvBlock(256,64)                # (128,128)
        s.out=nn.Sequential(
            nn.ConvTranspose2d(128,2,4,2,1),    # (256,256)
            nn.Tanh())
    def forward(s,x):
        e1=s.e1(x); e2=s.e2(e1); e3=s.e3(e2); e4=s.e4(e3)
        e5=s.e5(e4); e6=s.e6(e5); e7=s.e7(e6); e8=s.e8(e7)
        d1=s.d1(e8); d2=s.d2(torch.cat([d1,e7],1))
        d3=s.d3(torch.cat([d2,e6],1)); d4=s.d4(torch.cat([d3,e5],1))
        d5=s.d5(torch.cat([d4,e4],1)); d6=s.d6(torch.cat([d5,e3],1))
        d7=s.d7(torch.cat([d6,e2],1))
        return s.out(torch.cat([d7,e1],1))      # ab ∈ [-1,1]

# ---------- PatchGAN Discriminator ----------
class PatchD(nn.Module):
    def __init__(s):
        super().__init__()
        def C(in_c,out_c,norm=True):
            layers=[nn.Conv2d(in_c,out_c,4,2,1,bias=False)]
            if norm: layers.append(nn.BatchNorm2d(out_c))
            layers.append(nn.LeakyReLU(0.2,True))
            return nn.Sequential(*layers)
        s.model=nn.Sequential(
            C(3,64,norm=False),     # (128,128)
            C(64,128),              # (64,64)
            C(128,256),             # (32,32)
            C(256,512),             # (16,16)
            nn.Conv2d(512,1,4,1,1)  # (15,15) ≈ 70×70 patch
        )
    def forward(s,L,ab):
        x=torch.cat([L,ab],1)        # conditional
        return s.model(x)

G=UNetG().to(device)
D=PatchD().to(device)

# ---------- losses & optimisers ----------
criterion_GAN = nn.BCEWithLogitsLoss()
criterion_L1  = nn.L1Loss()
opt_G = optim.Adam(G.parameters(), LR_G, betas=(0.5,0.999))
opt_D = optim.Adam(D.parameters(), LR_D, betas=(0.5,0.999))
scaler_G = GradScaler(enabled=USE_AMP)
scaler_D = GradScaler(enabled=USE_AMP)

# ---------- training loop ----------
def train(batch_size=INIT_BATCH):
    train_loader=make_loader('train',batch_size)
    G.train(); D.train()
    for epoch in range(1,EPOCHS+1):
        g_loss_tot=d_loss_tot=0
        for L,ab in progress(train_loader,20,'Ep'+str(epoch)):
            L,ab=L.to(device),ab.to(device)
            real_label=torch.ones((L.size(0),1,15,15),device=device)
            fake_label=torch.zeros_like(real_label)

            # ---- 1. Train Discriminator ----
            opt_D.zero_grad(set_to_none=True)
            with amp():
                fake_ab = G(L)
                pred_real = D(L, ab)
                pred_fake = D(L, fake_ab.detach())
                loss_real = criterion_GAN(pred_real, real_label)
                loss_fake = criterion_GAN(pred_fake, fake_label)
                d_loss = (loss_real + loss_fake)*0.5
            scaler_D.scale(d_loss).backward()
            scaler_D.step(opt_D); scaler_D.update()
            d_loss_tot+=d_loss.item()

            # ---- 2. Train Generator ----
            opt_G.zero_grad(set_to_none=True)
            with amp():
                pred_fake = D(L, fake_ab)
                g_gan = criterion_GAN(pred_fake, real_label)
                g_l1  = criterion_L1(fake_ab, ab)
                g_loss = g_gan + λ_L1 * g_l1
            scaler_G.scale(g_loss).backward()
            scaler_G.step(opt_G); scaler_G.update()
            g_loss_tot+=g_loss.item()
        print(f'\nEpoch {epoch}/{EPOCHS}  G {g_loss_tot/len(train_loader):.4f}  D {d_loss_tot/len(train_loader):.4f}')
        torch.save(G.state_dict(), CKPT_G)
        torch.save(D.state_dict(), CKPT_D)

# ---------- auto-batch with OOM fallback ----------
def safe_train():
    bs=INIT_BATCH
    while True:
        try:
            train(bs); break
        except RuntimeError as e:
            if 'out of memory' in str(e).lower() and BATCH_AUTO and bs>1:
                bs//=2; print(f'\nOOM → retry with batch={bs}')
                if device.type=='cuda': torch.cuda.empty_cache()
            else: raise

if not CKPT_G.exists() or not CKPT_D.exists():
    safe_train()
else:
    print('Skipping training – checkpoints found.')

G.load_state_dict(torch.load(CKPT_G,map_location=device))
G.eval()

# ---------- optional CPU metrics ----------
if DO_METRICS:
    msssim=piq.MultiScaleSSIMLoss(data_range=1.)
    val_loader=make_loader('val',4)
    ms=0
    with torch.no_grad():
        for L,ab in progress(val_loader,40,'M'):
            fake_ab=G(L.to(device)).cpu()
            pred=lab_to_rgb_tensor(torch.cat([L,fake_ab],1))
            gt  =lab_to_rgb_tensor(torch.cat([L,ab],1))
            ms+=msssim(pred,gt).item()
    print(f'\nMS-SSIM (val): {ms/len(val_loader):.4f}')

# ---------- visual sanity-check ----------
if SHOW_IMAGES:
    val_loader=make_loader('val',1)
    L,ab=next(iter(val_loader))
    with torch.no_grad():
        fake_ab=G(L.to(device))
    pred=lab_to_rgb_tensor(torch.cat([L.to(device),fake_ab],1))[0].permute(1,2,0).cpu()
    gt  =lab_to_rgb_tensor(torch.cat([L,ab],1))[0].permute(1,2,0).cpu()
    gray=L[0][0].numpy()
    fig,ax=plt.subplots(1,3,figsize=(12,4))
    ax[0].imshow(gray,cmap='gray'); ax[0].set_title('Input'); ax[0].axis('off')
    ax[1].imshow(pred); ax[1].set_title('Predicted'); ax[1].axis('off')
    ax[2].imshow(gt);   ax[2].set_title('Ground truth'); ax[2].axis('off')
    plt.tight_layout(); plt.show()
```

```python
EPOCHS = 15
train()
```
